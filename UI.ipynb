```python
from tkinter import *
from functools import partial

def operation(Name_text, Married_text, Mobile_No_text, Applicant_Income_text, Loan_ID_text, Any_Dependent_text, Coapplicant_Income_text):
    
    name = "Name - " +Name_text.get()
    mob = "Mobile - " +Mobile_No_text.get()
    married = "Married Status - " +Married_text.get()
    app = "Applicant Income - " +Applicant_Income_text.get()
    loan = "Loan ID - " +Loan_ID_text.get()
    dep = "Dependent - " +Any_Dependent_text.get()
    coapp = "Co-applicant Income - " +Coapplicant_Income_text.get()
    
    n1 = Label(window, text=name).grid(row = 6, column = 1)
    n2 = Label(window, text=married).grid(row = 7, column = 1)
    n3 = Label(window, text=mob).grid(row = 8, column = 1)
    n4 = Label(window, text=app).grid(row = 9, column = 1)
    n5 = Label(window, text=loan).grid(row = 10, column = 1)
    n6 = Label(window, text=dep).grid(row = 11, column = 1)
    n7 = Label(window, text=coapp).grid(row = 12, column = 1)
    
    return

def model_check():
    check = Label(window, text="Congratulations!!! You are eligible and approved for this loan").grid(row = 7, column = 2)
    
def close():
    window.destroy()
    
window = Tk()
window.geometry('700x350')
window.title('Banking')

l1 = Label(window, text = 'Name').grid(row=0,column=0)
Name_text = StringVar()
e1 = Entry(window, textvariable=Name_text).grid(row=0,column=1)

l2 = Label(window, text = 'Mobile No.').grid(row=1,column=0)
Married_text = StringVar()
e2 = Entry(window, textvariable=Married_text).grid(row=0,column=3)

l3 = Label(window, text = 'Loan_ID').grid(row=2,column=0)
Mobile_No_text = StringVar()
e3 = Entry(window, textvariable=Mobile_No_text).grid(row=1,column=1)

l4 = Label(window, text = 'Married').grid(row=0,column=2)
Applicant_Income_text = StringVar()
e4 = Entry(window, textvariable=Applicant_Income_text).grid(row=1,column=3)

l5 = Label(window, text = 'Applicant Income').grid(row=1,column=2)
Loan_ID_text = StringVar()
e5 = Entry(window, textvariable=Loan_ID_text).grid(row=2,column=1)

l6 = Label(window, text = 'Any Dependent').grid(row=2,column=2)
Any_Dependent_text = StringVar()
e6 = Entry(window, textvariable=Any_Dependent_text).grid(row=2,column=3)

l7 = Label(window, text = 'Co-Applicant Income').grid(row=3,column=2)
Coapplicant_Income_text = StringVar()
e7 = Entry(window, textvariable=Coapplicant_Income_text).grid(row=3,column=3)

operation = partial(operation, Name_text, Married_text, Mobile_No_text, Applicant_Income_text, Loan_ID_text, Any_Dependent_text, Coapplicant_Income_text)
model_check = partial(model_check)
close = partial(close)
    
b1 = Button(window, text = "View", command = operation, width = 12).grid(row = 5, column=1)

b2 = Button(window, text = "Check", command = model_check, width = 12).grid(row =5 , column=2)

b3 = Button(window, text = "Close", command = close, width = 12).grid(row = 5, column=3)

window.mainloop()
```
